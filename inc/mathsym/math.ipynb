```python
def Text2St(ans,symbols_=["x","y","z","a","b","c"]):
    if ans.strip() == "": return "-3.1415926"
    dictionary = {'X': 'x', 'Y':'y', 'Z': 'z', '^': '**',"\\":"/"}
    transTable = ans.maketrans(dictionary)
    ans = ans.translate(transTable)
    ans = re.sub(r"X", r"x", ans)
    ans = re.sub(r"(\d)x", r"\1*x", ans)
    ans = ans = re.sub(r"[)][(]", r")*(", re.sub(r"(\d)x", r"\1*x", ans))
    ans=re.sub(r"x[(]", r"x*(", ans)
    ans = re.sub(r"x[ ]*", r"x", ans)
    ans = ans.replace(r"x^", r"x**")

    ans = re.sub(r"Y", r"y", ans)
    ans = re.sub(r"(\d)y", r"\1*y", ans)
    ans = ans = re.sub(r"[)][(]", r")*(", re.sub(r"(\d)y", r"\1*y", ans))
    ans=re.sub(r"y[(]", r"y*(", ans)
    ans = re.sub(r"y[ ]*", r"y", ans)
    ans = ans.replace(r"y^", r"y**")
    
    ans=re.sub(r"(\d)[(]", r"\1*(", ans)
    ans=re.sub(r"(\d)J[(]", r"\1*J(", ans)
    ans = ans.replace(r"J(", r"sqrt(")
    print(ans)
    ans = ans.replace(r"**", r"^")
    ans = re.sub(r"-[ ]*(\d+)[*]", r"+(0-\1)*",ans)
    ans = ans.replace(r"^", r"**")
    print(ans)
    return ans

def Put_PF201_Expr(TE):
    x=sp.Symbol('x')
    Val=TE["Val"]
    ans=TE["Ans"]
    if ans.strip() == "": ans = "3.1415926"
    ans=Text2St(ans)
    print("========")
    print(ans)
    print(sp.srepr(parse_expr(ans,transformations="all")))
    print(parse_expr(ans,transformations="all"))
    print(Val)
    print("========")
    try:
        if parse_expr(ans).subs({x:7})==Val.subs({x:7}):                   #比對答案:
            TE["OK"]=1
        else:                                      #不則
            TE["OK"]=0
    except:
        pass
        
a=random.choice([2,4,8,12,16,18,20,24,28,32,36,3,9,27,48,75,108,147])
b,c,d=np.random.choice([2,4,8,12,16,18,20,24,28,32,36,3,9,27,48,75,108,147],3)

# \sqrt{28} + \sqrt{9} - \sqrt{147}	2J(7)+3−7J(3)	-7*sqrt(3) + 3 + 2*sqrt(7)
a=28
b=9
c=147

st1=sp.sqrt(a)
st2=sp.sqrt(b)
st3=sp.sqrt(c)
St=st1+st2-st3
Stt=r" \sqrt{%s} + \sqrt{%s} - \sqrt{%s} "%(a,b,c)
display(St)
display(Stt)
Val=sp.simplify(St)

TE={"Val":Val,"Ans":"0-7","OK":0}
Put_PF201_Expr(TE)
TE={"Val":Val,"Ans":"0-7* sqrt(3)","OK":0}
Put_PF201_Expr(TE)
TE={"Val":Val,"Ans":"2J(7) + 3 - 7J(3)   ","OK":0}
Put_PF201_Expr(TE)
display(St)                

display(TE)
```
